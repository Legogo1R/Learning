Buble sort

```python
array = np.random.randint(10, size=10)
print(array)

def buble_sort(array_to_sort):
    
    array = array_to_sort
    lenght = len(array)
    while lenght > 0:
        i = 1

        while i < lenght:

            if array[i-1] >= array[i]:
                tmp_elem = array[i-1]
                array[i-1] = array[i]
                array[i] = tmp_elem

            i += 1

        lenght -= 1

    return array

buble_sort(array)
```

Selection sort

```python
array = np.random.randint(10, size=10)

print(array)

def selection_sort(array_to_sort):
    
    array = array_to_sort
    
    start_pos_bias = 0
    while start_pos_bias < len(array):

        i = start_pos_bias
        min_index = start_pos_bias

        while (i) < len(array):
            if array[min_index] >= array[i]:
                min_index = i
        
            i += 1

        tmp_elem = array[start_pos_bias]
        array[start_pos_bias] = array[min_index]
        array[min_index] = tmp_elem

        start_pos_bias += 1
        
    return array

selection_sort(array)
```

Insertion sort

```python
array = np.random.randint(10, size=10)
# array = [2,3,8,1,5]

print(array)

def insertion_sort(array_to_sort):
    
    array = array_to_sort
    
    i = 1

    while i < len(array):
        sorted = i-1

        while (sorted >= 0) and (array[sorted] > array[sorted+1]):
            tmp_elem = array[sorted]
            array[sorted] = array[sorted+1]
            array[sorted+1] = tmp_elem
            sorted -= 1
        
        i += 1
        
    return array

insertion_sort(array)
```

Binary search

```python
array = np.arange(2,13)

print(array)

def binary_search(array_to_sort, element):
    
    array = array_to_sort

    low = 0
    high = len(array) - 1
    selected_index = (low + high) // 2

    while array[selected_index] != element:
        selected_index = (low + high) // 2

        if array[selected_index] > element:
            high = selected_index - 1

        else:
            low = selected_index + 1
         
    return selected_index

binary_search(array, 12)
```

```python
array = np.arange(2,13)

print(array)

def recursive_binary_search(array_to_sort,
                            element,
                            low,
                            high):
    
    array = array_to_sort

    selected_index = (low + high) // 2

    if array[selected_index] > element:
        return recursive_binary_search(array_to_sort,
                                       element,
                                       low,
                                       selected_index - 1)

    elif array[selected_index] < element:
        return recursive_binary_search(array_to_sort,
                                       element,
                                       selected_index + 1,
                                       high)
         
    return selected_index

recursive_binary_search(array, 9, 0, len(array)-1)
```

Quick sort (Хоара)

```python
array = np.random.randint(10, size=10)
# array = [2,3,8,1,5,7,6,6,2,1,4]

print(array)

def quick_sort_Hoara_step(array_to_sort, left_pointer, right_pointer):
    
    array = array_to_sort
    pivot = array[(left_pointer + right_pointer) // 2]
    
    while left_pointer <= right_pointer:
        
        while array[left_pointer] < pivot:
            left_pointer += 1

        while array[right_pointer] > pivot:
            right_pointer -= 1

        if left_pointer <= right_pointer:

            tmp = array[left_pointer]
            array[left_pointer] = array[right_pointer]
            array[right_pointer] = tmp

            left_pointer += 1
            right_pointer -= 1

    return left_pointer

def quick_sort_Hoara_recursion(array_to_sort, start, end):

    if start >= end:
        return
    right_start = quick_sort_Hoara_step(array_to_sort, start, end)
    quick_sort_Hoara_recursion(array_to_sort, start, right_start-1)
    quick_sort_Hoara_recursion(array_to_sort, right_start, end)


def quick_sort_Hoara(array_to_sort):
    quick_sort_Hoara_recursion(array_to_sort, 0, len(array_to_sort) - 1)
    return array_to_sort


quick_sort_Hoara(array)
```

Merge sort

```python
array = list(np.random.randint(10, size=10))
# array = [2,3,8,1,5,7,6,6,2,1,4]

print(array)

def merge_sort(array_to_sort):
    
    array = array_to_sort

    if len(array) == 1:
        return array

    left_array = array[: (len(array) // 2)]
    right_array = array[(len(array) // 2) :]
    left = merge_sort(left_array)
    right = merge_sort(right_array)
    return merge_lists(left, right)

    
    

def merge_lists(left, right):
    sorted_list = []
    i = j = 0
    
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            sorted_list.append(left[i])
            i += 1
        else:
            sorted_list.append(right[j])
            j += 1

    if i < len(left):
        sorted_list += left[i:]
    if j < len(right):
        sorted_list += right[j:]

    return sorted_list



merge_sort(array)
```
